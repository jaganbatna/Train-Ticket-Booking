```python
# Line chart: KM driven over the years
plt.figure(figsize=(10,5))
plt.plot(df['year'], df['km_driven'], marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('KM Driven')
plt.title('KM Driven over the Years')
plt.grid(True)  # Add grid for better readability
plt.show()
```

```python
# Pie chart: Distribution of fuel types
plt.figure(figsize=(7,7))
plt.pie(df['fuel'].value_counts(), labels=df['fuel'].value_counts().index, autopct='%1.1f%%')
plt.title('Fuel Type Distribution')
plt.show()
```

```python
#DISTRIBUTION PLOT WITH KDE

plt.figure(figsize=(10, 6))
sns.displot(df['selling_price'], kind='kde')
plt.title('Distribution of Selling Price with KDE')
plt.xlabel('Selling Price')
plt.ylabel('Density')
plt.show()
```

```python
# Scatter plot: Mileage vs Selling price
plt.figure(figsize=(10,5))
plt.scatter(df['Mileage'], df['selling_price'])
plt.xlabel('Mileage')
plt.ylabel('Selling Price')
plt.title('Mileage vs Selling Price')
plt.show()
```

```python

# Function to detect outliers using the IQR method
def find_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter the dataframe to include only outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    return outliers

# Find outliers for each numerical column
outliers_dict = {}
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
for column in numerical_columns:
    outliers_dict[column] = find_outliers(df, column)

# Display the outliers found in each column
for column, outliers in outliers_dict.items():
    print(f"Outliers in {column}:")
    print(outliers)
    print("\n")
```

```python
df_sorted = df.sort_values('year')

# Plotting the line chart for 'year' vs 'selling_price'
plt.figure(figsize=(10,6))
plt.plot(df_sorted['year'], df_sorted['selling_price'], marker='o', linestyle='-', color='b')

# Add title and labels
plt.title('Year of Manufacture vs Selling Price', fontsize=14)
plt.xlabel('Year of Manufacture', fontsize=12)
plt.ylabel('Selling Price', fontsize=12)

# Display the chart
plt.grid(True)
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='fuel', y='selling_price', data=df)
plt.title('Fuel Type vs Selling Price')
plt.xlabel('Fuel Type')
plt.ylabel('Selling Price')

plt.show()
```

```python
#Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['selling_price'], bins=20, kde=True)
plt.title('Distribution of Selling Price')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='fuel', y='selling_price', data=df)
plt.title('Fuel Type vs Selling Price')
plt.xlabel('Fuel Type')
plt.ylabel('Selling Price')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='km_driven', y='selling_price', data=df)
plt.title('Scatter of KM Driven vs Selling Price')
plt.xlabel('KM Driven')
plt.ylabel('Selling Price')
plt.show()
```

```python
# Pair plot
sns.pairplot(df[numerical_columns])
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
corr = df[numerical_columns].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='owner', y='selling_price', data=df)
plt.title('Owner vs Selling Price')
plt.xlabel('Owner')
plt.ylabel('Selling Price')
plt.show()
```

```python
# prompt: DISTRIBUTION ASSESSMENT Q-Q PLOT

from scipy import stats
import matplotlib.pyplot as plt

# Assuming 'selling_price' is the column you want to assess
plt.figure(figsize=(8, 6))
stats.probplot(df['selling_price'], dist="norm", plot=plt)
plt.title("Q-Q Plot of Selling Price")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.show()
```

```python
# prompt: JOINT PLOT WITH KDE

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and 'selling_price' and 'km_driven' are relevant columns
plt.figure(figsize=(10, 6))
sns.jointplot(data=df, x='km_driven', y='selling_price', kind='kde')
plt.suptitle('Joint Plot of KM Driven vs Selling Price with KDE', y=1.02)
plt.show()
```
